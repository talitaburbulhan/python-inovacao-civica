```python
import pandas as pd
```

```python
dados_pe_rn_2020 = ['base_dados/consulta_cand_2020_PE.csv', 'base_dados/consulta_cand_2020_RN.csv']
```

```python
print(dados_pe_rn_2020)
```

```python
type(dados_pe_rn_2020)
```

```python
dataframes_pe_rn_2020 = []

for tabela in dados_pe_rn_2020:
    dataframe = pd.read_csv(tabela, sep=';', encoding='latin_1')
    dataframes_pe_rn_2020.append(dataframe)
```

```python
print(dataframes_pe_rn_2020)
```

```python
dataframe_completo = pd.concat(dataframes_pe_rn_2020, axis=0, ignore_index=True)
```
