```python
import pandas as pd
import glob
```

```python
list_arquivos = glob.glob('tarefa_TSE2022_notas-fiscais/nota*.csv')
```

```python
print(list_arquivos)
```

```python
type(list_arquivos)
```

```python
list_dfs = []

for item_list in list_arquivos:
    df = pd.read_csv(item_list, sep=';', encoding='latin_1')
    list_dfs.append(df)
    
```

```python
print(list_dfs)
```

```python
type(list_dfs)
```

```python
df_completo = pd.concat(list_dfs, axis=0, ignore_index=True)
```
